```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("sample.jpeg")
```

```python
img.shape
```

```python
rows,cols=img.shape[:2]
```

```python
plt.imshow(img)
```

```python
#CREATE TRANSFORMATION MATRIX
#FORMAT: [[1,0,tx],[0,1,ty]]
#tx-shift along X-AXIS (RIGHT)
#ty-shift along Y-AXIS (DOWN)

matrix=np.float32([[1,0,100],[0,1,50]])

#APPLY TRANSLATION
translated=cv2.warpAffine(img,matrix,(cols,rows))
plt.imshow(translated)
plt.show()
```

```python
#ROTATING THE IMAGE
#define center of rotation'
center=(cols//2,rows//2)
```

```python
#PARAMETERS: CENTER,ANGLE,SCALE
#angle - rotation in degrees (positive = counter-clockwise)
#scale - zoom (1=same size)
matrix=cv2.getRotationMatrix2D(center,-2,1)

rotated=cv2.warpAffine(img,matrix,(cols,rows))
plt.imshow(rotated)
plt.show()
```

```python
#RESIZE AN IMAGE
#PARAMETERS: (IMAGE, (WIDTH,HEIGHT))
resized=cv2.resize(img,(700,900))
plt.imshow(resized)
plt.show()
```

```python
#PERSPECTIVE TRANSFORM
#Concept: titled document -> top->down flat view
#format - define 4 source points (4 corners of document in original image)
#[top-left,top-right,bottom-right,bottom-left]
points1=np.float32([[100,150],[150,170],[550,550],[100,550]])

#define destination points
#CREATE A PERFECT RECTANGLE (FLAT VIEW)
points2=np.float32([[0,0],[300,0],[300,40],[0,400]])
matrix=cv2.getPerspectiveTransform(points1,points2)

#Apply perspective
warped = cv2.warpPerspective(img, matrix, (300,400))
plt.imshow(warped)
plt.show()
```
