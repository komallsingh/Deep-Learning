```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, SimpleRNN
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
text = [
    "i am learning deep learning",
    "i like deep learning",
    "deep learning is fun to learn",
    "i also like machine learning"
]
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(text)
```

```python
sequences = tokenizer.texts_to_sequences(text)
print(sequences)
```

```python
# Creating input and output pairs
X = []
y = []
for seq in sequences:
  for i in range(1, len(seq)):
    X.append(seq[:i])
    y.append(seq[i])
```

```python
# Pad sequences
X = pad_sequences(X)
y = np.array(y)
```

```python
vocab_size = len(tokenizer.word_index) + 1
```

```python

vocab_size
```

```python
# Build RNN Model
model = Sequential([
    Embedding(vocab_size, 10, input_length=X.shape[1]),
    SimpleRNN(32),
    Dense(vocab_size, activation='softmax')
])
```

```python
model.compile(loss="sparse_categorical_crossentropy",
              optimizer="adam")
```

```python
model.fit(X, y, epochs=100)
```

```python
input_text = "i am"
seq = tokenizer.texts_to_sequences([input_text])[0]
seq = pad_sequences([seq], maxlen=X.shape[1])
pred = model.predict(seq)
```

```python
pred
```

```python
pred_index = np.argmax(pred)
```

```python
for word, index in tokenizer.word_index.items():
  if index == pred_index:
    print("Next word :",word)
```
